## Data Analysis Report: Insights from Wrangled Twitter Data

#### Introduction:

In this report, I'm present a data analysis of the Twitter_archive_enhanced, tweet-json.txt, and image_predictions datasets that were previously cleaned and wrangled to address issues related to data quality and tidiness. The goal of this analysis was to gain insights into various aspects of Twitter usage and dog popularity.


* **Insight 1: What is the Most Popular Source of Tweets?**

The first insight I looked at the most popular source of tweets. To do this, I analyzed the source column in the Twitter_archive_enhanced dataset, which contains information about the device or application used to make the tweet.

I found that the majority of tweets came from the Twitter for iPhone source, accounting for over 2000 of all tweets. Other sources, such as Twitter Web Client and TweetDeck, made up the remaining tweets.

* **Insight 2: The Relationship between Favorite and Retweet Counts?**

The second insight I looked at the relationship between the number of favorites and retweets a tweet receives.

I found that there is a positive correlation between the two variables, with tweets that receive more favorites also receiving more retweets. This suggests that tweets that are more popular are also more likely to be shared by users.

* **Insight 3: What is the Popularity of Various Dog Types?*

The third insight I looked at the popularity of different dog types, as identified in the "dog_type" column of the Twitter_archive_enhanced dataset. This column was created by extracting the text from the 'doggo', 'floofer', 'pupper', 'puppo' columns, which contain information about the type of dog in a tweet.

I found that the most popular dog type was "pupper", accounting for nearly 66.4% of all tweets that contained a dog type. The second most popular dog type was "doggo", accounting for approximately 23.3% of tweets. The other two types, "floofer" and "puppo", were less popular.

* **Conclusion:**

In conclusion, our data analysis provided insight into various aspects of Twitter usage, including the most popular sources of tweets, the relationship between favorite and retweet counts, and the popularity of different dog types.
